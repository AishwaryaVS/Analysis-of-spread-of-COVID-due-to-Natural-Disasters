```python
import scipy
from scipy.integrate import odeint

import pandas as pd
import numpy as np
import datetime

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import datetime
import random
```

```python
data=pd.read_csv("/Users/aishwaryasheelvant/Downloads/covid-19-data-master/rolling-averages/us-states.csv")
```

```python
data=data.loc[(data.date>='2020-04-01') & (data.date<='2020-12-01')]
```

```python
data.info()
```

```python
data['date_week'] = pd.to_datetime(data['date']).apply(lambda x: x.isocalendar()[1])
```

```python
data
```

```python
df=data.groupby(['state','date_week']).mean().reset_index()
df
```

```python
states=['Arizona','Nevada','California','Oregon']
df_affected=df.loc[df['state'].isin(states)]
```

```python
df_affected_b=df_affected.loc[(df_affected['date_week']<=33) &(df_affected['date_week']>=29)]
# df_affected_b=df_affected_b.loc[(df_affected_b['date_week']<=45) &(df_affected_b['date_week']>40)]
```

```python
df_affected_b2=df_affected.loc[(df_affected['date_week']<=45) &(df_affected['date_week']>40)]
```

```python
import scipy.stats as stats
```

```python
df_affected_b
```

```python
stats.ttest_ind(a=df_affected_b2['cases'], b=df_affected_b['cases'], equal_var=True)
```

```python
df_unaffected=df.loc[~df['state'].isin(states)]
```

```python
df_unaffected
```

```python
df_unaffected_b=df_unaffected.loc[(df_unaffected['date_week']<=33) &(df_unaffected['date_week']>=29)]
df_unaffected_b2=df_unaffected.loc[(df_unaffected['date_week']<=45) &(df_unaffected['date_week']>40)]
```

```python
stats.ttest_ind(a=df_unaffected_b2['cases'], b=df_unaffected_b['cases'], equal_var=True)
```
